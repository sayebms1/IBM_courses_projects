## 1- County population

```python
import pandas as pd
```

```python
url = "https://www2.census.gov/programs-surveys/popest/datasets/2010-2019/counties/totals/co-est2019-alldata.csv"
county_pop = pd.read_csv(url,encoding='latin-1')
```

```python
county_pop = county_pop[county_pop['STNAME']=='Florida']
county_pop.reset_index(drop=True, inplace=True)
county_pop.head()
```

```python
col_not_2019=[col for col in county_pop.columns if '2019' not in col]
county_pop.drop(columns = col_not_2019, inplace=True )
```
